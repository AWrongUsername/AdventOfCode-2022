```python
def score_played(b):
    if b == 'X':
        return 1
    elif b == 'Y':
        return 2
    else:
        return 3
```

```python
def score_outcome(a, b):
    if (a == b):
        return 3
    elif (a == 'rock' and b == 'scissors'):
        return 0
    elif (a == 'paper' and b == 'rock'):
        return 0
    elif (a == 'scissors' and b == 'paper'):
        return 0
    else:
        return 6
```

```python
file = open('input.txt', 'r')
lines = file.readlines()
lines = list(map(lambda x:x.strip(),lines))
lines
```

```python
scores = {
    'A': 'rock',
    'B': 'paper',
    'C': 'scissors',
    'X': 'rock',
    'Y': 'paper',
    'Z': 'scissors'
}
```

```python
total = 0

for row in lines:
    a, b = row.split()
    total += score_outcome(scores.get(a), scores.get(b))
    total += score_played(b)

total
```
