```python
def score_played(b):
    if b == 'X':
        return 1
    elif b == 'Y':
        return 2
    else:
        return 3
```

```python
def score_outcome(a, b):
    if (a == b):
        return 3
    elif (a == 'rock' and b == 'scissors'):
        return 0
    elif (a == 'paper' and b == 'rock'):
        return 0
    elif (a == 'scissors' and b == 'paper'):
        return 0
    else:
        return 6
```

```python
win = {'A': 'Y', 'B': 'Z', 'C': 'X'}
lose = {'A': 'Z', 'B': 'X', 'C': 'Y'}
draw = {'A': 'X', 'B': 'Y', 'C': 'Z'}


def should_play(a, b):
    # need to lose
    if (b == 'X'):
        return lose[a]
    # Need to draw
    if(b == 'Y'):
        return draw[a]
    # Need to win
    if(b == 'Z'):
        return win[a]
```

```python
file = open('input.txt', 'r')
lines = file.readlines()
lines = list(map(lambda x:x.strip(),lines))
lines
```

```python
scores = {
    'A': 'rock',
    'B': 'paper',
    'C': 'scissors',
    'X': 'rock',
    'Y': 'paper',
    'Z': 'scissors'
}
```

```python
total = 0

for row in lines:
    a, b = row.split()
    play = should_play(a, b)

    total += score_played(play)
    total += score_outcome(scores.get(a), scores.get(play))

print(total)
```
